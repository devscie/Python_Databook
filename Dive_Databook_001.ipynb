Possível também analisar a coluna de dose, obtemos 2 tipos de doses diferentes, **d1** e **d2**.

Obtido o retorno de **```dados['droga'].unique()```** , séries de códigos como resposta. Esse tipo de dados 'caracteres' possivelmente foram criptografados para anonimizar os tipos de drogas administradas, para tentar evitar qualquer tipo de viés na análise dos resultados.

* Há uma série de regras quando alguns experimentos, testes e pesquisas são feitas, proteger os dados como nome, sexo, e outros fatores explicitos, dependendo da análise a ser feita, para que sejam evitados vieses.

Próxima analise, coluna nomeada como: **g-0**

````python
dados['g-0'].unique()
````

Conhecendo a origem da base de dados, seu contexto e a que se refere os dados, conseguimos entender que essa letra **'g'** da coluna, remete à palavra **gene**. Ou seja, esses números nos dizem a expressão de cada gene referente a exposição as drogas.

* Nesta etapa da análise, já conseguimos identificar e entender as informações de diferentes colunas; as colunas que nos indicam os tratamentos, as drogas, o tempo, e depois as respostas genéticas, que tem a letra g no ínicio do seu nome.

A profundando mais na análise podemos entender a distribuição dessas informações contidas no dataframe. 

* Por exemplo, quantos experimentos temos utilizando droga e quantos não utilizam;

* Quantos usam o d1, e quantos utilizam o d2, e assim por diante.

````python
dados['tratamento'].value_counts()
````

Utilizamos o **`.value_counts`** já que o objetivo é classificar os tipos e totalizar as quantidades de valores que contem nas colunas, nesse caso, na **coluna tratamento**.

Obtemos o retornou 2 linhas:

* Uma com a quantidade de elementos que temos, na categoria **com_droga** e a outra com a quantidade que temos na categoria **com_controle**.

Houve uma grande diferença no resultado dessas categorias. 

````python
dados['dose'].value_counts()
````

Nessa **coluna dose** os valores demonstram mais equilibradas. Mas somente com o número é difícil fazermos uma leitura mais aprofundada.

Para entendermos melhor e fazer um comparativo, necessario traçarmos a proporção entre os dados, para isso setamos como parâmetro, dentro dos parênteses o **`normalize = True :`**


````python
dados['tratamento'].value_counts(normalize = True)
````

Obtemos os dados **"normalizados"**. Podemos interpretar utilizando a porcentagem (multiplicando por 100), o que nos daria algo em torno de **92%** versus **8%** (aproximadamente), mostrando o tamanho da desproporção.

Repetimos o mesmo tratamento com a **coluna dose**:

````python
dados['dose'].value_counts(normalize = True)
````

Para representar essas informações visualmente, temos uma série de gráficos. E, geralmente, em muito casos é usado o gráfico de pizza.

````python
dados['tratamento'].value_counts().plot.pie()
````

Em azul, temos a quantidade de **tratamento com drogas** (em maior proporção), e em laranja temos o **tratamento "com controle"**.

````python
dados['tempo'].value_counts().plot.pie()
````

Ao analisar a **coluna tempo** atráves do gráfico de pizza ficou mais difícil compará as porporções.

Esse tipo de gráfico dificulta a análise, especialmente quando as informações e valores estão melhor balanceadas.

> *É recomendado ao fazer algum tipo de análise, evitar o uso de gráficos que nos remetem ao nome de comida, como por exemplo: rosca, pizza, torta, e assim por diante.*

Para melhorarmos a visualização, podemos utilizar o gráfico de barras.


````python
dados['tempo'].value_counts().plot.bar()
````

Com o gráfico de barras conseguimos identificar com mais facilidade qual observação tem a sua maior frequência. 

* Ao contrario do que nota no gráfico de pizza, a quantidade de horas com o maior número de observações foi a de 48.

* No eixo y temos o número que nos remete ao número de observações. Então podemos observar que tivemos pouco mais de 8000 observações relativas às 48 horas.

Constatamos, que o gráfico de barras demostra ser mais útil, nesse caso, do que o gráfico de pizza.

No dataframe, **coluna g-0**, temos alguns valores dentro de um intervalo definido. Para não termos valores muito distantes entre si, é bastante comum no meio científico que haja uma normalização dos resultados, para criamos um intervalo que não seja tão grande, em que o meio dessa distribuição seja o 0.

````python
dados_filtrados = dados[dados['g-0'] > 0]
dados_filtrados.head()
````

Dessa maneira, **temos somente as 5 primeiras linhas com o valores maiores do que 0 na coluna g-0**, com a ajuda dessa máscara com o código **`[dados['g-0']>0]`**.

Da mesma forma que aplicamos essa máscara, podemos também utilizar **'querys'**.
